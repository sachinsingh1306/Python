```python
# Write the program to find out wether a student is pass or fail. Assume 3 subject and total will be 40% and in each subject aleast 33%
# marks should have to pass take marks from user input.

marks1 = int(input("Enter Subject 1 makrs : "))
marks2 = int(input("Enter Subject 2 makrs : "))
marks3 = int(input("Enter Subject 3 makrs : "))

total_percentage = (100*(marks1 + marks2 + marks3)/300)
if(marks1>= 33 and marks2>= 33 and marks3>= 33 and total_percentage >= 40):
    print(f"You are passed with {total_percentage}%")
else:
    print(f"You are failed with {total_percentage}%")
```

```python
# A spam comment is define as a text containing following keywords. 
# make a lot of money" , "buy now" , "subscribe now" , "click this".
# write the program to detect the spams.

a = "make a lot of money"
b = "buy now"
c = "click this"
d = "subscribe now"

enter = input("Enter a message...")
message = enter.lower()

if( a in message or b in message or c in message or d in message):
    print(f"This is a spam message : {enter}")
else:
    print(enter)
```

```python
# Write a program to calculate the grade of the student fro his marks.

marks = int(input("Enter your marks out of 100: "))

if marks > 100 or marks < 0:
    print("Invalid marks! Please enter between 0 and 100.")

elif marks > 90:
    print("Congratulations! Your grade is Ex")
elif marks >= 80:
    print("Congratulations! Your grade is A")
elif marks >= 70:
    print("Your grade is B. Passed!")
elif marks >= 60:
    print("Your grade is C. Passed!")
elif marks >= 50:
    print("Your grade is D. Passed!")
else:
    print("You are Fail! Your Grade is F.")

```

```python
# Q4. Age Group Checker
# Ask the user for their age.
# age < 13 → “Child”
# 13–19 → “Teenager”
# 20–59 → “Adult”
# 60+ → “Senior Citizen”

age = int(input("Enter your age : "))

if age >= 0 and age < 13:
    print("Child")
elif age >= 13 and age <= 19:
    print("Teenager")
elif age >= 20 and age <= 59:
    print("Adult")
elif age >= 60:
    print("Senior Citizen")
else:
    print("Invalid age")
```

```python
# Taking first number input
num1 = int(input("Enter Number 1st : "))

# Taking second number input
num2 = int(input("Enter Number 2nd : "))

# Taking operator (+, -, *, /)
operator = input("Enter operator : ")

if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    print(num1 / num2)
else:
    print("Invalid Operator")
```
