```python
# ðŸ”¹ 2. Function that Returns Filtered Names

# Example: Return names starting with S

def names_start_with_s(names):
    result = []
    for name in names:
        if name.startswith("S"):
            result.append(name)
    return result

names = ["Sachin", "Rohan", "Sunil", "Amit"]
print(names_start_with_s(names))
```

```python
# ðŸ”¹ 3. Function that Counts Words Containing a Letter

def count_with_a(words):
    count = 0
    for word in words:
        if "a" in word.lower():
            count += 1
    return count

words = ["apple", "pen", "banana"]
print(count_with_a(words))
```

```python
# Write a function that:
# Takes a list of names
# Returns a list of names whose length is greater than 4

def names_with_length(names):
    result = []
    for name in names:
        if len(name) > 4:
            result.append(name)
    return result

names = ["Sachin", "Rohan", "Amit"]
print(names_with_length(names))
```

```python
# Write a function that:
# Takes a list of words
# Returns how many words start with a vowel

def words_start_with(words):
    count = 0
    for word in words:
        if word[0].lower() in "aeiou":
            count += 1
    return count

words = ["apple", "banana", "orange"]
print(words_start_with(words))

```

```python
# Write a function that:
# Takes a list of sentences
# Returns the total count of the word "python" (case-insensitive)

def find_words(sentences):
    count = 0
    for sentence in sentences:
        count += sentence.lower().count("python")
    return count

sentences = ["Python is easy", "I love python"]

print(find_words(sentences))
        
            
```

```python
# Write a function that:
# Takes a list of names
# Returns a dictionary with:
# Key = name
# Value = length of the name
names = ["Sachin", "Rohan", "Amit"]

def name_len_dict(names):
    result = {}
    for name in names:
        result[name] = len(name)
    return result

names = ["Sachin", "Rohan", "Amit"]
print(name_len_dict(names))
```

```python
# Write a function that:
# Takes a list of numbers
# Returns a new list containing only even numbers

def new_list_even_number(numbers):
    newList = []
    for number in  numbers:
        if number % 2 == 0:
            newList.append(number)
    return newList
numbers = [1, 2, 3, 4, 6]
print(new_list_even_number(numbers))
```

```python
'''Practice Question 6
Write a function that:
Takes a list of words
Returns a list of words that are longer than 5 characters'''

def new_list_of_words(words):
    new_list = []
    for word in words:
        if len(word) > 5:
            new_list.append(word)
    return new_list

words = ["apple", "banana", "grapefruit"]

print(new_list_of_words(words))
```
